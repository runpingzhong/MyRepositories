```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error,mean_absolute_error
#plt.rcParams["font.family"] = 'Arial Unicode MS'
#plt.rcParams['axes.unicode_minus'] = False
```

```python
traing=pd.read_excel('样本数据.xlsx',index_col=0)
traing.head()
data=pd.read_excel('优化之后各个样本数据.xlsx',index_col=0).reset_index(drop=True)
data.head()
```

```python
data.columns=traing.columns
```

```python
#读取因变量
y=pd.read_excel('/Users/zhongrunping/Desktop/jupyter文件夹/数学建模/RON损失(不是变量).xlsx')[::-1].reset_index(drop=True)
y=y.iloc[:,2]
```

```python
#训练模型
svr=SVR(kernel='poly',coef0=5,degree=3,cache_size=800).fit(traing,y)
mean_absolute_error(y,svr.predict(traing))
#训练平均的绝对误差0.15
```

```python
#预测优化后的
y_after=svr.predict(data)
```

```python
(y-y_after).mean()
#平均降低了损失值0.69
```

```python
#平均损失降幅55%
((y-y_after).mean())/y.mean()
```

```python
loss=pd.DataFrame(y_after,columns=['损失值'])
```

```python
#将损失值从小到大排序
loss.sort_values(by='损失值')
```

```python
sorted_loss=data.iloc[loss.sort_values(by='损失值').index,:]
```

```python
sorted_loss=sorted_loss.reset_index()
```

```python
sorted_loss['index']=sorted_loss['index']+1
```

```python
sorted_loss.columns
```

```python
sorted_loss.columns=['样本编号', '原料换热器管程总管进口温度', '1.0MPa蒸汽进装置温度', '氮气进装置流量', '稳定塔顶压力',
       'K101机出口压力', '蒸汽进装置流量', '加氢裂化轻石脑油进装置流量', '辛烷值1RON', '硫含量2,μg/g']
```
